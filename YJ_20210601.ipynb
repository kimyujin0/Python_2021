```python
# 상관관계와 공분산
import pandas_datareader.data as web
import pandas as pd
```

```python
all_data = {ticker: web.get_data_yahoo(ticker)
            for ticker in ['AAPL','IBM','MSFT','GOOG']}
price = pd.DataFrame({ticker: data['Adj Close']
                      for ticker, data in all_data.items()})
volume = pd.DataFrame({ticker: data['Volume']
                       for ticker, data in all_data.items()})
```

```python
# h2는 많으니까 반복문
for found in html_soup.find_all('h2'):
    print(found.text)
```

```python
episodes = []
ep_tables = html_soup.find_all('table', class_ = 'wikitable plainrowheaders wikiepisodetable')
ep_tables
```

```python
for table in ep_tables:
    headers=[]
    rows = table.find_all('tr')
    for header in table.find("tr").find_all("th"):
        headers.append(header.text)
        
    for row in table.find_all('tr')[1:]:
        values = []
        for col in row.find_all(['th','td']):
            values.append(col.text)
        if values:
            episode_dict = {headers[i]:values[i] for i in range(len(values))}
            episodes.append(episode_dict)
            
for episode in episodes:
    print(episode)
```

```python
df = pd.DataFrame(episodes)
df.head()
```

```python
import sqlite3
```

```python
# SQL 데이터
query = "CREATE TABLE test(a VARCHAR(20), b VARCHAR(20),c REAL, d INTEGER);" # 20글자를 넘지 않는 a라는 칸 만들기
con = sqlite3.connect('mydata.sqlite')
con.execute(query) # 만들어진 말을 database server에게 전달해야함
```

```python
con.commit() #commit : 데이터베이스에서 영구적인 변경을 확정하는 일(취소가 안됨)
```

```python
data = [('atlanta', 'georgia',1.25,6), ('Tallahassee', 'Florida', 2.6,3), ('Sacramento', 'California',1.7,5)]
query = "INSERT INTO test VALUES(?,?,?,?)"
con.executemany(query,data)
con.commit()
```

```python
query = "SELECT * FROM test"
cur = con.execute(query)
rows = cur.fetchall()   # 3줄, 모든 패치를 가져옴
rows
```
