```python
# 산점도와 색상넣는 3가지 방법
dataA = np.random.standard_normal((100,2))
dataB = np.random.standard_normal((100,2))
dataC = np.random.standard_normal((100,2))
dataA += np.array((-3,-3))
dataB += np.array((4,4))
dataC += np.array((-1,4))
plt.scatter(dataA[:,0], dataA[:,1], color = "b")
plt.scatter(dataB[:,0], dataB[:,1], color = "0.5") # 0.5=회색, 0~1    !가장선호!
plt.scatter(dataC[:,0], dataC[:,1], color = "#D1B2FF")  # 색상팔레트 사용 가능
plt.show()
```

```python
# 산점도와 색상넣는 3가지 방법
dataA = np.random.standard_normal((100,2))
dataB = np.random.standard_normal((100,2))
dataC = np.random.standard_normal((100,2))
dataA += np.array((-3,-3))
dataB += np.array((4,4))
dataC += np.array((-1,4))
plt.scatter(dataA[:,0], dataA[:,1], edgecolor="0", color = "b")  # edgecolor = 테두리
plt.scatter(dataB[:,0], dataB[:,1], edgecolor="0", color = "0.5") # 0.5=회색, 0~1    !가장선호!
plt.scatter(dataC[:,0], dataC[:,1], edgecolor="0", color = "#D1B2FF")  # 색상팔레트 사용 가능
plt.show()
# 겹치는 곳에서 구분을 잘 할 수 있음
```

```python
import matplotlib.cm as cm
N = 256
angle = np.linspace(0, 8*2*np.pi, N)
radius = np.linspace(5, 1, N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X,Y, c=angle, cmap=cm.Spectral)
plt.show()
```

```python
X = np.linspace(-10,10,100)  # X는 하나만 있어도 됨
Y1 = [2*x**2+1 for x in X]
Y2 = [3*x**2+5 for x in X]
Y3 = [0.5*x**2+3 for x in X]
plt.plot(X,Y1, linestyle="solid") # ㅡ
plt.plot(X,Y2, linestyle="dashed") # ---
plt.plot(X,Y3, linestyle="dashdot") # -.-.-
plt.show()
```

```python
# 누적막대그래프
N= 4
A = np.random.random(N)
B = np.random.random(N)
X = np.arange(N)
plt.bar(X,A,color="red")
plt.bar(X,A+B, bottom=A)
plt.show()
```

```python
# 세가지 변수의 누적막대그래프
N= 4
A = np.array([1,2,3,2])
B = np.array([2,3,2,1])
C = np.array([1,4,2,3])
X = np.arange(N)
plt.bar(X,A,color="red")
plt.bar(X,A+B,color="green", bottom=A)
plt.bar(X,A+B+C, bottom=A+B, color="blue")
plt.show()
```

```python
# hatch를 적용하여 막대그래프에 빗금 그리기
N= 4
A = np.array([1,2,3,2])
B = np.array([2,3,2,1])
C = np.array([1,4,2,3])
X = np.arange(N)
plt.bar(X,A,color="red", hatch='/')
plt.bar(X,A+B,color="green", hatch= 'x', bottom=A)
plt.bar(X,A+B+C, bottom=A+B, color="blue", hatch='/')
plt.show()
```

```python
# 색깔이 같을 때 모양(marker)으로 구별하기
A = np.random.standard_normal((100,2))
B = np.random.standard_normal((100,2))
A += np.array((-1,-2))
B += np.array((0.5,1))
plt.scatter(A[:,0], A[:,1], color='k',marker="o")
plt.scatter(B[:,0], B[:,1], color='k',marker='.')
# marker종류 : ^(삼각형),oXx,v(아래삼각형),>(오른쪽 삼각형),<(왼쪽삼각형),s(사각형),p(오각형),P(+),*,D(다이아몬드),'',.
plt.show()
```

```python
# 싱크함수그래프
X = np.linspace(-6,6,1024)
Y1 = np.sinc(X)
Y2 = np.sinc(X)+1
plt.plot(X,Y1, marker='o', color='0.7')
plt.plot(X,Y2, marker='o', color='0.7', markevery=32)   # markevery =n : n개마다 하나씩 점찍음
plt.show()
```

```python
# 여러가지 시각화
M = np.random.standard_normal((500,2))
R = np.sum(M**2, axis=1)
plt.ylim(-5,5) # 글씨적을 공간을 만들어줌
plt.xlim(-5,5)
plt.scatter(M[:,0],M[:,1], c='0.7', marker='s', s=32*R, edgecolor='k')
plt.show()
```

```python
M = np.random.standard_normal((500,2))
R = np.sum(M**2, axis=1)
plt.figure(figsize=(5,5))  # 좌표평면 크기 지정
plt.ylim(-5,5)
plt.xlim(-5,5)
plt.scatter(M[:,0],M[:,1], c='0.7', marker='s', s=32*R, edgecolor='k')
plt.show()
```

```python
# 작은화면 만들기
X = np.linspace(-6,6,1024)
X_small = np.linspace(-3,3,1024)
Y_small = np.sinc(X_small)
rect = plt.Rectangle((-3,-0.2),6, 1)
plt.add_patch(rect)
plt.plot(X, np.sinc(X), c='0.7')
sub_axes = plt.axes([0.6,0.6,0.25,0.25])
sub_axes.plot(X_small,Y_small, c='0.0')
plt.show()
```

```python
# 극좌표계
T = np.linspace(0, 2*np.pi, 1024)
plt.axes(polar=True)
plt.plot(T, 1+0.25*np.sin(16*T))
plt.show()
```

```python
# 극좌표계
import matplotlib.patches as patches
plt.axes(polar=True)
radius = [1,0.5,0.7,0.3,0.1,0.2,0.6,0.8,0.5]  # 개수는 자유롭게 찍을 수 있음
theta = np.linspace(0, 2*np.pi, len(radius), endpoint=False)
points = np.vstack((theta, radius)).T
plt.gca().add_patch(patches.Polygon(points, color='0.5'))
plt.show()
```
